```python
class Student:
    def __init__(self,name,age,email):
        self.name=name
        self.age=age
        self.email=email
        
    def studentDetail(self):
        print(f"Name: {self.name} , Age:{self.age} , Email:{self.email}")


name=input("Enter Name:")
age=input("Enter Age:")
email=input("Enter Email:")
 
std=Student(name,age,email)
std.studentDetail()

```

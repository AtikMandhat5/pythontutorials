```python
 class Student:
        a=input('Enter name of the student1:')
        b=input('Enter age of the student1:')
        c=input('Enter degree of student1:')
        d=input('Enter name of the student2:')
        e=input('Enter age of the student2:')
        f=input('Enter degree of student2:')

        def std1(self):
            print('Stud_1.name:',self.a)
            print('Stud_1.age:',self.b)
            print('Stud_1.graduate:',self.c)

        def std2(self):
            print('Stud_2.name:',self.d)
            print('Stud_2.age:',self.e)
            print('Stud_2.graduate:',self.f)

Stud_1 = Student()
Stud_2 = Student()

Stud_1.std1()
Stud_2.std2()





```

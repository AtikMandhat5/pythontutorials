```python
import numpy as np
```

```python
mxt =np.matrix("4,5,16,7;2,-3,2,3;3,4,5,6;4,7,8,9")
print(mxt)
```

```python
det_matrix=np.linalg.det(mxt)
```

```python
print(det_matrix)
```

```python

mxt1 =np.matrix("4,5,16,7,5; 2,-3,2,3,4; 3,4,5,6,8; 4,7,8,9,3; 7,5,9,6,7")
print(mxt1)
rank_matrix=np.linalg.matrix_rank(mxt1)
print(rank_matrix)
```
